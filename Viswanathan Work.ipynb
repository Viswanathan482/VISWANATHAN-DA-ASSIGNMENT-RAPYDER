# Top 5 locations for Top 20 Industries

```python
top5locations = pd.DataFrame({"Location": data.Location.str.split(",", expand = True)[0],"Industry": data.Industry.str.split(",", expand = True)[0]})
top5locations.dropna(inplace = True)
```

```python
top5cities = pd.crosstab(top5locations.Location, top5locations.Industry).reset_index()
top5cities = top5cities.melt(id_vars = "Location", value_name = "Job_Opportunities")
top5cities = top5cities.sort_values(by = ["Industry", "Job_Opportunities"], ascending = False).reset_index(drop = True)
top5cities = top5cities.dropna()
top5cities = top5cities.sort_values(by = "Job_Opportunities", ascending = False).reset_index(drop = True)

top20industries = top5cities.groupby("Industry").sum().reset_index().sort_values(by = "Job_Opportunities", ascending = False)[:20]
```

```python
import plotly.express as px
```

```python
# saved industry name in list for fetching only 5 same name industry or only five Location of particular indusry

industry20 = top20industries.Industry.reset_index(drop = True)
limit = 0
something = pd.DataFrame(columns=["Location", "Industry", "Job_Opportunities"])
for j in range(20):
    limit = 0
    for i in range (top5cities.size):
        if limit == 5:
            break
        if industry20[j] == top5cities.loc[i, "Industry"]:
            something = something.append(top5cities.loc[i], ignore_index = True)
            limit += 1
```

```python
# plot pie chart of top 5 location of top 20 Industry
for i in industry20:
    top20industries = something.loc[something.Industry == i, ["Location", "Job_Opportunities"]]
    fig = px.pie(top20industries, names = "Location", values = "Job_Opportunities", color = "Location",
                 title = "Top 5 Location for " + i + " Industry")
    fig.show()
```

# Skills Required for various Jobs


```python
skills = data[["Role Category", "Key Skills"]].copy()
classy = skills['Key Skills'].str.split("|", expand = True) # split keys
classy = pd.concat([skills, classy], axis = 1) # concat splited keys columns
classy = classy.melt(id_vars = ["Role Category", "Key Skills"], var_name = "Job_Opportunities", value_name = "Skills")
# transform Key skills using melt
classy = classy.dropna() 
classy.Job_Opportunities = 1 # assign job opportunites value 1 for counting purpose
classy = classy.groupby(["Role Category", "Skills"]).Job_Opportunities.sum().reset_index()
# group by role and skills and find sum of job opportunities
classy = classy.sort_values(by = "Job_Opportunities", ascending = False).reset_index(drop = True)
```

```python
skillsjobcat = pd.DataFrame(columns = ["Role Category", "Skills", "Job_Opportunities"])
rolecategory = top_20_JobCategory.Role_Category
limit = 0
for i in range(20):
    limit = 0
    for j in range(classy.size):
        if limit == 10:
            break
        if rolecategory[i] == classy.loc[j, "Role Category"]:
            skillsjobcat = skillsjobcat.append(classy.loc[j])
            limit += 1
```

# Average salary offered in Top 10 cities

```python
salary_loc = data[["Job Salary", "Location"]].copy()
salary_loc["New_Location"] = salary_loc.Location.str.strip().str.split(",", expand = True)[0]

classy = salary_loc['Job Salary'].apply(lambda x: re.sub("[^0-9 -]", "", str(x)))
salary_loc["Min_Salary"] = classy.str.strip().str.split("-", expand = True)[0]
salary_loc["Max_Salary"] = classy.str.strip().str.split("-", expand = True)[1]
salary_loc.dropna(inplace = True)

# remove space between digits
salary_loc.Min_Salary = salary_loc.Min_Salary.str.replace(" ", "")
salary_loc.Max_Salary = salary_loc.Max_Salary.str.replace(" ", "")

# put 0 where value is totally empty
salary_loc.loc[salary_loc.Max_Salary == "", "Max_Salary"] = "0"
salary_loc.loc[salary_loc.Min_Salary == "", "Min_Salary"] = "0"

# convert it into 
salary_loc.Min_Salary = salary_loc.Min_Salary.astype("int64")
salary_loc.Max_Salary = salary_loc.Max_Salary.astype("int64")
salary_loc.drop(columns = ["Location", "Job Salary"], inplace = True) # drop old columns
```

```python
income = salary_loc.groupby("New_Location").median().reset_index() # find average of salary

toplocation_salary = pd.DataFrame()
loc_list = top_20_cities.Location
for i in range(20):
    for j in range(income.New_Location.size):
        if loc_list[i] == income.loc[j, "New_Location"]:
            toplocation_salary = toplocation_salary.append(income.loc[j])
            break

toplocation_salary = toplocation_salary.reset_index(drop = True)
toplocation_salary = toplocation_salary[:10].copy() # find top 10 
toplocation_salary.rename(columns = {"New_Location": "Location"}, inplace = True)
toplocation_salary.Max_Salary = toplocation_salary.Max_Salary.astype("int64")
toplocation_salary.Min_Salary = toplocation_salary.Min_Salary.astype("int64")
toplocation_salary = toplocation_salary[toplocation_salary.columns[[2, 1, 0]]] # change column order
```

```python
plt.figure(figsize = (12, 10))
ax = plt.subplot(111)
ax.bar(toplocation_salary.index - 0.2, toplocation_salary.Min_Salary, color = "pink", width = 0.4, label = "Min Salary")
ax.bar(toplocation_salary.index +0.2, toplocation_salary.Max_Salary, color = "purple", width = 0.4, label = "Max Salary")
plt.xlabel("Location")
plt.ylabel("Average Salary")
plt.xticks([i for i in toplocation_salary.index] ,toplocation_salary.Location)
plt.legend()
plt.show()
```

# Average salary offered by the top 10 Industries in the top 5 locations


```python
new_avg_sal_ind = data[["Location", "Industry", "Job Salary"]].copy() # preparing data
new_avg_sal_ind.Location = new_avg_sal_ind.Location.str.split(",", expand = True)[0] # preparing location
new_avg_sal_ind.Industry = new_avg_sal_ind.Industry.str.split(",", expand = True)[0] # preparing industry
new_avg_sal_ind['Job Salary'] = new_avg_sal_ind['Job Salary'].apply(lambda x: re.sub("[^0-9 -]", "", str(x))) # salary
new_avg_sal_ind["Min_Salary"] = new_avg_sal_ind['Job Salary'].str.split("-", expand = True)[0] # min salary
new_avg_sal_ind["Max_Salary"] = new_avg_sal_ind['Job Salary'].str.split("-", expand = True)[1] # max salary
new_avg_sal_ind.Min_Salary = new_avg_sal_ind.Min_Salary.str.replace(" ", "")
new_avg_sal_ind.Max_Salary = new_avg_sal_ind.Max_Salary.str.replace(" ", "")
new_avg_sal_ind.loc[new_avg_sal_ind.Min_Salary == "", "Min_Salary"] = "0"
new_avg_sal_ind.loc[new_avg_sal_ind.Max_Salary == "", "Max_Salary"] = "0"
new_avg_sal_ind.dropna(inplace = True)
new_avg_sal_ind.Min_Salary = new_avg_sal_ind.Min_Salary.astype("int64")
new_avg_sal_ind.Max_Salary = new_avg_sal_ind.Max_Salary.astype("int64")
```

```python
classy = new_avg_sal_ind.groupby(["Industry", "Location"]).median().reset_index()
# group industry and location and find average of salary
classy.Industry = classy.Industry.str.upper() # convert upper case of industry
lst_Industry = Industries.Industry[:25] # get 25 top industry
lst_Location = top_20_cities.Location # get 20 top industry
limit = 0
top_10_avg_sal_industry_location = pd.DataFrame(columns = ["Industry", "Location", "Max_Salary", "Min_Salary"])
# creating empty data frame for storing filtering data
for i in range(25):
    limit = 0
    dummy_df = classy.loc[classy.Industry == lst_Industry[i]].reset_index(drop = True)
    # creating dummy df for top 10 industry
    for j in range(20):
        if limit == 5:
            break
        else:
            for k in range(dummy_df.Location.size):
                if lst_Location[j] == dummy_df.loc[k, "Location"]:
                    # append top 5 country and it's average salary of top 10 industry
                    top_10_avg_sal_industry_location = top_10_avg_sal_industry_location.append(dummy_df.loc[k])
                    limit += 1
                    break
```

```python
classy = top_10_avg_sal_industry_location.melt(id_vars = ["Industry", "Location"], var_name = "Salaries")
px.sunburst(data_frame = classy, path = ["Industry", "Location", "Salaries"],values = "value", title = "Average salary of top 10 Industry in top 5 Location", height = 600)
```
